```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import association_rules, apriori
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_excel(r"C:\Users\lenovo\OneDrive\Desktop\Daily needs\DS Assingments\Done\Association Rules\Association Rules\Online retail.xlsx")
df
```

# Data Preprocessing

```python
df.head()
```

```python
df.isna().sum() # There are no null values present in the dataset 
```

```python
transactions = df.values.tolist() # here we convert the dataset values into the python's list of list 
transactions = [','.join(transaction) for transaction in transactions] # by here we join it using (,)
transactions = [transaction.split(',') for transaction in transactions] # here we split the dataset to be identified by the python
```

```python
#Here we perform encoding technique to convert the rows into unique values of the items present 
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()
```

```python
te_df= te.fit(transactions).transform(transactions)
```

```python
df_encoded = pd.DataFrame(te_df, columns=te.columns_)
```

```python
df_encoded.head()
```

# Implementation of Apriori Algorithm 

```python
frequent_itemsets = apriori(df_encoded, min_support=0.01, use_colnames=True)
frequent_itemsets
```

# Generating Association rules over the dataset processed using Apriori Algorithm

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.2)
rules = rules[(rules['support'] >= 0.02) & 
              (rules['confidence'] >= 0.3) & 
              (rules['lift'] >= 1.2)]
rules = rules[['antecedents', 'consequents', 'antecedent support', 'consequent support','support', 'confidence', 'lift']]
rules
```

```python
# Sort by lift for most interesting rules
rules.sort_values(by='lift', ascending=False).head(10)
```

# What is support and Confidence. How do you calculate them?
- Support: raio of transactions within antecendent and consequent itemsets that appears in a data
- Formula to calulate Support: (Antecedent+Consequent)/Total transactions
- Confidence: ratio of antecedent(if) transctions that also have the consequent(then) item set
- Formula to calulate Confidence: (Antecedent+consequent)/Antecedent

# What are some limitations or challenges of Association rules mining?
